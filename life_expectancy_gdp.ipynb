```python
### LOADING LIBRARIES ###
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
### LOADING DATA ###
df = pd.read_csv("all_data.csv")
```

```python
### CLEANING DATA ###
df = df.rename(columns = {"Country":"country", "Year":"year", "Life expectancy at birth (years)":"life_expect", "GDP":"gdp"})
df["gdp_b"] = df.gdp/1000000000
```

```python
### Creating a 3x2 set of subplots that creates a scatter plot for each country and their relationship of GDP and Life Expectancy ###
sns.set_style("ticks")
fig, ax = plt.subplots(2,3, figsize=(20,10))
ax[0,0].scatter(data=df[df.country == "Chile"], x="life_expect", y="gdp_b", c="steelblue", s=50)
ax[0,0].set_title("Chile", fontsize=15)
ax[0,1].scatter(data=df[df.country == "China"], x="life_expect", y="gdp_b", c="mediumseagreen", s=50)
ax[0,1].set_title("China", fontsize=15)
ax[0,2].scatter(data=df[df.country == "Germany"], x="life_expect", y="gdp_b", c="chocolate", s=50)
ax[0,2].set_title("Germany", fontsize=15)
ax[1,0].scatter(data=df[df.country == "Mexico"], x="life_expect", y="gdp_b", c="orchid", s=50)
ax[1,0].set_title("Mexico", fontsize=15)
ax[1,1].scatter(data=df[df.country == "United States of America"], x="life_expect", y="gdp_b", c="gold", s=50)
ax[1,1].set_title("United States of America", fontsize=15)
ax[1,2].scatter(data=df[df.country == "Zimbabwe"], x="life_expect", y="gdp_b", c="lightskyblue", s=50)
ax[1,2].set_title("Zimbabwe", fontsize=15)
sns.despine()


fig.add_subplot(111, frame_on=False)
plt.tick_params(labelcolor="none", bottom=False, left=False)
plt.xlabel("Life Expectancy at Birth (Years)", fontsize=18, labelpad=10)
plt.ylabel("GDP (Billions $)", fontsize=18, labelpad=20)
plt.title("Life Expectancy and GDP", fontsize=30, pad=30)

labels= ["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"]
fig.legend(labels=labels, loc="right", fontsize=20)
plt.subplots_adjust(right=0.85)
plt.savefig("Country_LifeExpectGDP.png")
plt.show()
```

```python
### Creating a single plot that shows all of the information together to get a better idea of how the information relates to one another###
sns.set_style("whitegrid")
fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.scatter(data=df[df.country == "Chile"], x="life_expect", y="gdp_b", c="steelblue", s=50)
ax1.scatter(data=df[df.country == "China"], x="life_expect", y="gdp_b", c="mediumseagreen", s=50)
ax1.scatter(data=df[df.country == "Germany"], x="life_expect", y="gdp_b", c="chocolate", s=50)
ax1.scatter(data=df[df.country == "Mexico"], x="life_expect", y="gdp_b", c="orchid", s=50)
ax1.scatter(data=df[df.country == "United States of America"], x="life_expect", y="gdp_b", c="gold", s=50)
ax1.scatter(data=df[df.country == "Zimbabwe"], x="life_expect", y="gdp_b", c="lightskyblue", s=50)

plt.xlabel("Life Expectancy at Birth (Years)", fontsize=15, labelpad=10)
plt.ylabel("GDP (Billions $)", fontsize=15, labelpad=20)
plt.title("Life Expectancy and GDP", fontsize=20, pad=30)

labels= ["Chile", "China", "Germany", "Mexico", "USA", "Zimbabwe"]
fig1.legend(labels=labels, fontsize=15, bbox_to_anchor=(.4, .8))

plt.savefig("One_LifeExpectGDP.png")
plt.show()
```
